```python
#Importing the required libs for the exercise

from os import path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from ydata_synthetic.synthesizers import ModelParameters
from ydata_synthetic.preprocessing.timeseries import processed_stock
from ydata_synthetic.synthesizers.timeseries import TimeGAN
```

```python
#Specific to TimeGANs
seq_len=24
n_seq = 6
hidden_dim=24
gamma=1

noise_dim = 32
dim = 128
batch_size = 128

log_step = 100
learning_rate = 5e-4

gan_args = ModelParameters(batch_size=batch_size,
                           lr=learning_rate,
                           noise_dim=noise_dim,
                           layers_dim=dim)
```

```python
quote = 'TATASTEEL.NS'
df = pd.read_csv(''+quote+'.csv')
df
```

```python
df = df.drop(['Date'], axis=1)
```

```python
#Feature Scaling
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler(feature_range=(0,1))#Scaled values btween 0,1
training_set_scaled=sc.fit_transform(df)
#In scaling, fit_transform for training, transform for test
```

```python
training_set_scaled.shape
```

```python
scaled_data = np.array(training_set_scaled)
scaled_data = np.reshape(scaled_data, (scaled_data.shape[0],scaled_data.shape[1],1))#.shape 0=row,1=col
print(scaled_data.shape)
```

```python
if path.exists('synthesizer_stock.pkl'):
    synth = TimeGAN.load('synthesizer_stock.pkl')
else:
    synth = TimeGAN(model_parameters=gan_args, hidden_dim=24, seq_len=seq_len, n_seq=n_seq, gamma=1)
    synth.train(scaled_data, train_steps=1000)
    synth.save('synthesizer_stock.pkl')
```

```python
synth_data = synth.sample(len(scaled_data))
print(synth_data.shape)
```

```python
synth_data
```

```python
new_arr = synth_data[:746, 0, :]
new_arr
```

```python
gandata = sc.inverse_transform(new_arr)
gandata
```

```python
# assuming new_arr is your NumPy array
gan_df = pd.DataFrame(gandata.reshape(-1, gandata.shape[-1]))

# the above line reshapes your array to (n_samples, 6) and converts it to a DataFrame
# if you want to give column names to your DataFrame, you can do so as follows:
gan_df.columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
gan_df.to_csv('Gan_Stock_Data.csv', index=False)
gan_df
```

```python
# to set the plot size
plt.figure(figsize=(16, 8), dpi=150)
  
# using plot method to plot open prices.
# in plot method we set the label and color of the curve.
df['Close'].plot(label='Real Data', color='orange')
gan_df['Close'].plot(label='Generated Data')

  
# adding title to the plot
plt.title('Close Price Plot')
  
# adding Label to the x-axis
plt.xlabel('Years')
  
# adding legend to the curve
plt.legend()
```
