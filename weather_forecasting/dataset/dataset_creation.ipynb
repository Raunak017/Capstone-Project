```python
# import required modules
import zipfile
import pandas as pd

l1 = ['_2016a','_2016b','_2017a','_2017b','_2018a','_2018b','_2019a','_2019b','_2020a','_2020b','_2021a','_2021b','_2022a','']
n = 'mpi_roof'
l2 = ['2016a','2016b','2017a']
df = pd.DataFrame()
# open zipped dataset
for i in l1:
    with zipfile.ZipFile(n + i + ".zip") as z:
        # open the csv file in the dataset
        with z.open(n + i + ".csv") as f:
            f1 = pd.read_csv(f , encoding= 'unicode_escape')
            df = df.append(f1, ignore_index=True)
            
                
```

```python
compression_opts = dict(method='zip',
                        archive_name='final_dataset.csv')  
df.to_csv('final_dataset.zip', compression=compression_opts, index=False)
```
